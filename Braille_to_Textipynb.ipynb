```python
# ============================================
# Google Colab: Unicode Braille PDF вЖТ English
# FULL Correct Translator (Capitals, Numbers, Punctuation)
# ============================================

!pip install pymupdf

import fitz  # PyMuPDF
from google.colab import files

# --------------------------------------------
# Upload PDF
# --------------------------------------------
print("Upload your Braille Unicode PDF:")
uploaded = files.upload()

pdf_file = list(uploaded.keys())[0]
print("Uploaded:", pdf_file)


# --------------------------------------------
# Extract Braille Unicode Text from PDF
# --------------------------------------------
def extract_braille_text(pdf_path):
    doc = fitz.open(pdf_path)
    braille_text = ""

    for page in doc:
        braille_text += page.get_text()

    return braille_text


braille_text = extract_braille_text(pdf_file)

print("\n===== BRAILLE TEXT EXTRACTED =====\n")
print(braille_text[:500], "...")  # preview only


# --------------------------------------------
# Braille Unicode Grade 1 Translator
# --------------------------------------------

# Letters a-z
BRAILLE_ALPHA = {
    "в†Б":"a","в†Г":"b","в†Й":"c","в†Щ":"d","в†С":"e",
    "в†Л":"f","в†Ы":"g","в†У":"h","в†К":"i","в†Ъ":"j",
    "в†Е":"k","в†З":"l","в†Н":"m","в†Э":"n","в†Х":"o",
    "в†П":"p","в†Я":"q","в†Ч":"r","в†О":"s","в†Ю":"t",
    "в†•":"u","в†І":"v","в†Ї":"w","в†≠":"x","в†љ":"y","в†µ":"z"
}

# Digits (after number sign в†Љ)
BRAILLE_NUMBERS = {
    "в†Б":"1","в†Г":"2","в†Й":"3","в†Щ":"4","в†С":"5",
    "в†Л":"6","в†Ы":"7","в†У":"8","в†К":"9","в†Ъ":"0"
}

# Punctuation
BRAILLE_PUNCT = {
    "в†В":",",
    "в†≤":".",
    "в†¶":"вАЬ",
    "в†і":"вАЭ",
    "в†Ц":"!",
    "в†ґ":"?",
    "в†§":"-",
    "в†Р":":",
    "в†Ж":";",
    "в†А":" "   # Braille space
}

CAPITAL_SIGN = "в††"
NUMBER_SIGN  = "в†Љ"


# --------------------------------------------
# Full Translator Function
# --------------------------------------------
def braille_to_english(text):

    result = ""
    capitalize_next = False
    number_mode = False

    for ch in text:

        # Capital indicator
        if ch == CAPITAL_SIGN:
            capitalize_next = True
            continue

        # Number indicator
        if ch == NUMBER_SIGN:
            number_mode = True
            continue

        # Space resets number mode
        if ch == "в†А":
            number_mode = False
            result += " "
            continue

        # Number mode translation
        if number_mode and ch in BRAILLE_NUMBERS:
            result += BRAILLE_NUMBERS[ch]
            continue

        # Letter translation
        if ch in BRAILLE_ALPHA:
            letter = BRAILLE_ALPHA[ch]

            if capitalize_next:
                letter = letter.upper()
                capitalize_next = False

            result += letter
            continue

        # Punctuation translation
        if ch in BRAILLE_PUNCT:
            result += BRAILLE_PUNCT[ch]
            continue

        # Unknown symbol
        result += ch

    return result


# --------------------------------------------
# Convert Braille вЖТ English
# --------------------------------------------
english_text = braille_to_english(braille_text)

print("\n\n===== FULL ENGLISH OUTPUT =====\n")
print(english_text)


```

```python
import re

def clean_wrapped_lines(text):
    # Fix broken words caused by PDF wrapping
    text = re.sub(r"(\w+)\n(\w+)", r"\1 \2", text)
    return text


def split_question_answer(text):
    text = text.strip()

    # Split at first blank line
    parts = re.split(r"\n\s*\n", text, maxsplit=1)

    if len(parts) == 2:
        question = clean_wrapped_lines(parts[0]).strip()
        answer = clean_wrapped_lines(parts[1]).strip()
        return question, answer

    return text, ""


# Apply
question, answer = split_question_answer(english_text)

print("\n==============================")
print("        QUESTION")
print("==============================\n")
print(question)

print("\n==============================")
print("        ANSWER")
print("==============================\n")
print(answer)
```
